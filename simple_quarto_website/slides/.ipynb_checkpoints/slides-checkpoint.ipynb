## Code with its output plot 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 50, 200)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
