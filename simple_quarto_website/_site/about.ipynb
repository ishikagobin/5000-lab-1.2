```python
import numpy as np
import matplotlib.pyplot as plt

# simple example plot
x = np.linspace(0,50, 400)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Plot")
plt.xlabel("x")
plt.ylabel("value")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
